```python
import numpy as np
import librosa
from keras.models import load_model

def predict_audio_class(file_path, model_path, max_pad_len=100, segment_duration=1.0, short_threshold=5, num_segments_long=3):
    """
    Predict the class of an audio file using a trained model. For short files, use the first segment(s).
    For long files, use segments from beginning, middle, and end.

    Parameters:
        file_path (str): Path to the audio file to predict.
        model_path (str): Path to the trained model.
        max_pad_len (int): Maximum padding length for MFCC features.
        segment_duration (float): Duration of each segment (in seconds).
        short_threshold (float): Threshold duration for classifying a file as short (in seconds).
        num_segments_long (int): Number of segments to use for long files.

    Returns:
        str: Predicted class label.
    """
    # Load trained model
    model = load_model(model_path)

    # Load audio file
    audio, sample_rate = librosa.load(file_path, sr=None)

    # Get the duration of the audio in seconds
    audio_duration = librosa.get_duration(y=audio, sr=sample_rate)

    # Split the audio into fixed segments
    segment_samples = int(segment_duration * sample_rate)
    total_samples = len(audio)
    segments = [audio[i:i + segment_samples] for i in range(0, total_samples, segment_samples)]

    def extract_mfcc_features(audio_segment, sr, max_pad_len=100):
        """Extract MFCC features from an audio segment."""
        mfccs = librosa.feature.mfcc(y=audio_segment, sr=sr, n_mfcc=40)
        pad_width = max_pad_len - mfccs.shape[1]
        if pad_width > 0:
            mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
        else:
            mfccs = mfccs[:, :max_pad_len]
        return mfccs.T  # Transpose to have the shape (max_pad_len, n_mfcc)

    note_predictions = []
    prediction_confidences = []  # Store confidence for weighted voting

    # Handle short files by predicting based on the first few segments
    if audio_duration <= short_threshold:
        print(f"Processing as short file (duration: {audio_duration}s)")
        for segment in segments[:2]:  # Use the first 2 segments if available
            features = extract_mfcc_features(segment, sample_rate, max_pad_len)
            features_reshaped = features[np.newaxis, ..., np.newaxis]
            prediction = model.predict(features_reshaped)
            predicted_class = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction)
            note_predictions.append(predicted_class)
            prediction_confidences.append(confidence)

    # Handle long files by predicting based on the beginning, middle, and end segments
    else:
        num_segments = len(segments)
        # Select 3 segments: one from the beginning, middle, and end
        print(f"Processing as long file (duration: {audio_duration}s)")
        if num_segments >= num_segments_long:
            selected_segments = [
                segments[0],  # Beginning
                segments[num_segments // 2],  # Middle
                segments[-1]  # End
            ]
        else:
            # In case of very few segments, just use all of them
            selected_segments = segments
        # Predict for each selected segment and store predictions with confidence
        for segment in selected_segments:
            features = extract_mfcc_features(segment, sample_rate, max_pad_len)
            features_reshaped = features[np.newaxis, ..., np.newaxis]
            prediction = model.predict(features_reshaped)
            predicted_class = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction)
            note_predictions.append(predicted_class)
            prediction_confidences.append(confidence)

    # Ensure that note_predictions is not empty
    if not note_predictions:
        raise ValueError("No predictions were made. Check if the audio segments were correctly processed.")

    # Weighted majority voting based on confidence
    weighted_votes = np.zeros(2)  # We have two classes
    for i, prediction in enumerate(note_predictions):
        weighted_votes[prediction] += prediction_confidences[i]

    # Get the class with the highest confidence
    predicted_class = np.argmax(weighted_votes)

    # Map numerical class to instrument names
    class_names = ['flute', 'gataberaya']
    predicted_label = class_names[predicted_class]

    return predicted_label

# Example usage
model_path = '/content/drive/MyDrive/my_model.h5'
unseen_audio_path = '/content/drive/MyDrive/Unseen/New-Recording-203.wav'

predicted_label = predict_audio_class(unseen_audio_path, model_path)
print(f"The predicted class for the unseen audio file is: {predicted_label}")
```
