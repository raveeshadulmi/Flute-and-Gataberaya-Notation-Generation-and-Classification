```python
import numpy as np
import librosa
from keras.models import load_model

def predict_audio_class(file_path, model_path, max_pad_len=100, segment_duration=1.0, short_threshold=5, num_segments_long=3):
    """
    Predict the class of an audio file using a trained model. For short files, use the first segment(s).
    For long files, use segments from beginning, middle, and end.

    Parameters:
        file_path (str): Path to the audio file to predict.
        model_path (str): Path to the trained model.
        max_pad_len (int): Maximum padding length for MFCC features.
        segment_duration (float): Duration of each segment (in seconds).
        short_threshold (float): Threshold duration for classifying a file as short (in seconds).
        num_segments_long (int): Number of segments to use for long files.

    Returns:
        str: Predicted class label.
    """
    # Load trained model
    model = load_model(model_path)

    # Load audio file
    audio, sample_rate = librosa.load(file_path, sr=None)

    # Get the duration of the audio in seconds
    audio_duration = librosa.get_duration(y=audio, sr=sample_rate)

    # Split the audio into fixed segments
    segment_samples = int(segment_duration * sample_rate)
    total_samples = len(audio)
    segments = [audio[i:i + segment_samples] for i in range(0, total_samples, segment_samples)]

    def extract_mfcc_features(audio_segment, sr, max_pad_len=100):
        """Extract MFCC features from an audio segment."""
        mfccs = librosa.feature.mfcc(y=audio_segment, sr=sr, n_mfcc=40)
        pad_width = max_pad_len - mfccs.shape[1]
        if pad_width > 0:
            mfccs = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)), mode='constant')
        else:
            mfccs = mfccs[:, :max_pad_len]
        return mfccs.T  # Transpose to have the shape (max_pad_len, n_mfcc)

    note_predictions = []
    prediction_confidences = []  # Store confidence for weighted voting

    # Handle short files by predicting based on the first few segments
    if audio_duration <= short_threshold:
        # Use the first segment (or first few segments)
        print(f"Processing as short file (duration: {audio_duration}s)")
        for segment in segments[:2]:  # Use the first 2 segments if available
            features = extract_mfcc_features(segment, sample_rate, max_pad_len)
            features_reshaped = features[np.newaxis, ..., np.newaxis]  # Add batch and channel dimensions
            prediction = model.predict(features_reshaped)
            predicted_class = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction)
            note_predictions.append(predicted_class)
            prediction_confidences.append(confidence)

    # Handle long files by predicting based on the beginning, middle, and end segments
    else:
        num_segments = len(segments)

        # Select 3 segments: one from the beginning, middle, and end
        print(f"Processing as long file (duration: {audio_duration}s)")
        if num_segments >= num_segments_long:
            selected_segments = [
                segments[0],  # Beginning
                segments[num_segments // 2],  # Middle
                segments[-1]  # End
            ]
        else:
            # In case of very few segments, just use all of them
            selected_segments = segments

        # Predict for each selected segment and store predictions with confidence
        for segment in selected_segments:
            features = extract_mfcc_features(segment, sample_rate, max_pad_len)
            features_reshaped = features[np.newaxis, ..., np.newaxis]
            prediction = model.predict(features_reshaped)
            predicted_class = np.argmax(prediction, axis=1)[0]
            confidence = np.max(prediction)
            note_predictions.append(predicted_class)
            prediction_confidences.append(confidence)

    # Ensure that note_predictions is not empty
    if not note_predictions:
        raise ValueError("No predictions were made. Check if the audio segments were correctly processed.")

    # Weighted majority voting based on confidence
    weighted_votes = np.zeros(2)  # We have two classes
    for i, prediction in enumerate(note_predictions):
        weighted_votes[prediction] += prediction_confidences[i]

    # Get the class with the highest confidence
    predicted_class = np.argmax(weighted_votes)

    # Map numerical class to instrument names
    class_names = ['flute', 'gataberaya']
    predicted_label = class_names[predicted_class]

    return predicted_label

# Example usage
model_path = '/content/drive/MyDrive/my_model.h5'
unseen_audio_path = '/content/drive/MyDrive/Unseen/New-Recording-203.wav'

predicted_label = predict_audio_class(unseen_audio_path, model_path)
print(f"The predicted class for the unseen audio file is: {predicted_label}")
```

```python
import librosa
import numpy as np
import joblib
import google.generativeai as genai
from scipy.io import wavfile
import scipy.fft
import scipy.fftpack

# Set up the API key and configure the Generative AI model (outside if-else)
GOOGLE_API_KEY = 'GOOGLE_API_KEY'
genai.configure(api_key=GOOGLE_API_KEY)
genai_model = genai.GenerativeModel('gemini-pro')


def notation_generation(predicted_label, unseen_audio_path, notes):

      # Only execute the following code if the predicted label is 'flute'
      if predicted_label == 'flute':

          # Find the note and octave corresponding to a frequency
          def get_note(frequency, notes):
              for note, freq in notes:
                  if freq * 0.986 <= frequency <= freq * 1.015:
                      return note
              return ''

          # Frequency detection function (Matching the second version's working logic)
          def detect_notes(threshold=0.01, min_freq=100, max_freq=10000, amplitude_threshold=0.02):
              # Read the audio file
              sample_rate, signal = wavfile.read(unseen_audio_path)

              # Convert stereo to mono if necessary
              if len(signal.shape) == 2:
                  signal = signal.sum(axis=1) / 2

              N = signal.shape[0]
              time_samples_per_sec = 1.0 / sample_rate

              # FFT computation
              fft = np.abs(scipy.fft.fft(signal))
              fft_frequencies = scipy.fftpack.fftfreq(N, time_samples_per_sec)
              fft_one_side = fft[:N // 2]  # Use only the positive frequencies
              fft_frequencies_one_side = fft_frequencies[:N // 2]

              normalized_amplitude = fft_one_side / np.linalg.norm(fft_one_side)
              recorded_notes = []

              # Note detection (This part is essential to match the second version)
              for i, freq in enumerate(fft_frequencies_one_side):
                  if min_freq <= freq <= max_freq:
                      amplitude = normalized_amplitude[i]
                      if amplitude > amplitude_threshold:
                          note = get_note(freq, notes)
                          if note and note not in recorded_notes:
                              recorded_notes.append(note)

              return recorded_notes

          # Call the detect_notes function with adjusted parameters
          detected_notes = detect_notes(threshold=0.01, min_freq=110, max_freq=8200, amplitude_threshold=0.02)
          print("Detected notes:", detected_notes)

          # Prepare the prompt for the generative AI model
          prompt = f"""You are a music teacher with knowledge in both oriental music in Sinhala and western notations.
          Below is how the notes are written in Sinhala:

          C4: ස4
          C#4: කෝමල රි4
          D4: රි4
          D#4: කෝමල ග4
          E4: ග4
          F4: ම4
          F#4: තිව්ර ම4
          G4: ප4
          G#4: කෝමල ධ4
          A4: ධ4
          A#4: කෝමල නි4
          B4: නි4
          C5: ස5

          For the following musical notes, generate the Sinhala musical notations:

          {', '.join(detected_notes)}
          """

        # Generate the response from the model
          response = genai_model.generate_content(prompt)

          # Print the response
          print(response.text)

      # If the predicted label is 'gataberaya'
      elif predicted_label == 'gataberaya':

          # Load the KNN and SVM models and label encoders
          knn_label_encoder_path = '/content/drive/MyDrive/GataberayaClassification/label_encoder.pkl'
          knn_label_encoder = joblib.load(knn_label_encoder_path)
          knn_model_path = '/content/drive/MyDrive/g_knn_1.pkl'
          knn_model = joblib.load(knn_model_path)

          # Load the trained SVM model and label encoder (for audio longer than 6 seconds)
          svm_model_path = '/content/drive/MyDrive/g_svm.pkl'  # Update this path if needed
          svm_model = joblib.load(svm_model_path)
          svm_label_encoder_path = '/content/drive/MyDrive/label_encoder1.pkl'
          svm_label_encoder = joblib.load(svm_label_encoder_path)

          # Function to load and preprocess the unseen audio file for a given segment
          def preprocess_unseen_audio(file_path, fixed_length=80):
              # Load the audio file
              audio, sr = librosa.load(file_path, sr=None)

              # Define function to pad or truncate features
              def pad_or_truncate(feature, fixed_length):
                  if feature.shape[1] < fixed_length:
                      padding = np.zeros((feature.shape[0], fixed_length - feature.shape[1]))
                      feature = np.concatenate((feature, padding), axis=1)
                  else:
                      feature = feature[:, :fixed_length]
                  return feature

              # Extract features
              mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)
              mfcc = pad_or_truncate(mfcc, fixed_length)

              chroma = librosa.feature.chroma_stft(y=audio, sr=sr)
              chroma = pad_or_truncate(chroma, fixed_length)

              spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr)
              spectral_contrast = pad_or_truncate(spectral_contrast, fixed_length)

              zcr = librosa.feature.zero_crossing_rate(y=audio)
              zcr = pad_or_truncate(zcr, fixed_length)

              tonnetz = librosa.feature.tonnetz(y=librosa.effects.harmonic(y=audio), sr=sr)
              tonnetz = pad_or_truncate(tonnetz, fixed_length)

              rmse = librosa.feature.rms(y=audio)
              rmse = pad_or_truncate(rmse, fixed_length)

              # Flatten and concatenate all features
              combined_features = np.hstack([
                  mfcc.flatten(),
                  chroma.flatten(),
                  spectral_contrast.flatten(),
                  zcr.flatten(),
                  tonnetz.flatten(),
                  rmse.flatten()
              ]).reshape(1, -1)  # Reshape to match the input format

              return combined_features

          # Function to predict notes using the appropriate model (KNN or SVM)
          def predict_unseen_audio(file_path):
              # Load and preprocess the audio file
              combined_features = preprocess_unseen_audio(file_path)

              # Determine whether to use KNN or SVM based on audio duration
              audio, sr = librosa.load(file_path, sr=None)
              audio_duration = librosa.get_duration(y=audio, sr=sr)
              if audio_duration > 3:
                  print(f"Using SVM model for audio duration {audio_duration} seconds")
                  # Use SVM model for audio longer than 3 seconds
                  model = svm_model
                  label_encoder = svm_label_encoder
              else:
                  print(f"Using KNN model for audio duration {audio_duration} seconds")
                  # Use KNN model for audio shorter than or equal to 3 seconds
                  model = knn_model
                  label_encoder = knn_label_encoder

              # Predict the notes (classes) of the unseen audio file
              prediction = model.predict(combined_features)
              predicted_class = label_encoder.inverse_transform(prediction)  # Use the appropriate label encoder

              return predicted_class.tolist()

          # Predict the notes for the unseen audio file
          detected_notes = predict_unseen_audio(unseen_audio_path)

          # Print the detected notes
          print(f"Detected notes for the unseen audio file: {detected_notes}")

          # Prepare the prompt dynamically with the detected notes
          prompt = f"""You are a music teacher with knowledge in both oriental music in Sinhala and western notations.
          Below is how the notes are written in Sinhala:

          thath = තත්
          jith = ජිත්
          dom = දොම්
          nang = නන්

          For the following musical notes, generate the Sinhala musical notations:

          {', '.join(detected_notes)}
          """

          # Generate the response from the generative AI model
          response = genai_model.generate_content(prompt)
          print(response.text)


predicted_notes = notation_generation(predicted_label, unseen_audio_path, notes)
```
